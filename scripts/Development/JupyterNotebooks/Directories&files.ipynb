```python
import os
import csv
```

```python
def main(dirName,opFile):
    listOfFiles = list()
    for (dirpath, dirnames, filenames) in os.walk(dirName):
        listOfFiles += [os.path.join(dirpath, file) for file in filenames]
            
    with open(opFile, "w") as outfile:
        for elem in listOfFiles:
            outfile.write(elem)
            outfile.write("\n")
```

```python
dirName = '//mxns.loc//shares//CoE//Lab Operations//Nutria//ITdev//Install_Kits//2019'
opFile = '//mxns.loc//shares//CoE//Lab Operations//Nutria//ITdev//Install_Kits//2019_Files.csv'
main(dirName,opFile)
```
